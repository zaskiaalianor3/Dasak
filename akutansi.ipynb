```python
import pandas as pd

# List untuk menyimpan data transaksi
stream_transaksi = []

print("=== Aplikasi Akuntansi Sederhana (Debit–Kredit) ===")
print("Masukkan transaksi satu per satu. Ketik 'selesai' untuk berhenti.\n")

while True:
    akun = input("Nama akun (atau ketik 'selesai'): ")

    if akun.lower() == "selesai":
        break
    
    jenis = input("Jenis (debit/kredit): ").lower()
    while jenis not in ["debit", "kredit"]:
        print("Input salah! Ketik 'debit' atau 'kredit'")
        jenis = input("Jenis (debit/kredit): ").lower()
    
    try:
        jumlah = float(input("Jumlah (Rp): "))
    except:
        print("Jumlah harus angka!")
        continue
    
    # Tambah ke stream transaksi
    stream_transaksi.append({
        "Akun": akun,
        "Jenis": jenis.capitalize(),
        "Jumlah": jumlah
    })
    
    print("Transaksi ditambahkan.\n")


# Membuat DataFrame
df = pd.DataFrame(stream_transaksi)

print("\n=== Tabel Transaksi ===")
display(df)

# Hitung total debit dan kredit
total_debit = df[df["Jenis"] == "Debit"]["Jumlah"].sum()
total_kredit = df[df["Jenis"] == "Kredit"]["Jumlah"].sum()

print("\nTotal Debit  : Rp", total_debit)
print("Total Kredit : Rp", total_kredit)

# Cek apakah seimbang
if total_debit == total_kredit:
    print("\nStatus: ✔ Jurnal SEIMBANG")
else:
    print("\nStatus: ✘ Jurnal TIDAK seimbang (Balance error)")
```
